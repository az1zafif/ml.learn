```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns
sns.set()
```

```python
rnd_num = np.random.RandomState(42)
X = np.dot(rnd_num.rand(2,2), rnd_num.randn(2,500)).T
```

```python
#X
```

```python
X[:,0] = - X[:,0]
```

```python
plt.scatter(X[:,0], X[:,1]);
plt.axis('equal');
```

# Principal Components Identification

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
print(pca.components_) #Object array multidimens, merupakan himpunan semua vektor eigen
```

```python
print(pca.explained_variance_)#perbedaan
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.3)
```

```python
#plot data
for k, v in zip(pca.explained_variance_, pca.components_):
    vec = v*3* np.sqrt(k)
    
    ax = plt.gca()
    arrowprops = dict(arrowstyle= '<-',
                                   linewidth=1, shrinkA=0, shrinkB=0)
    ax.annotate('',pca.mean_, pca.mean_ + vec, arrowprops=arrowprops)
    ax.text(-0.90,1.2,'PC1', ha = 'center', rotation=42,size=12)
    ax.text(-0.1,-0.6,'PC2', ha = 'center', rotation=50,size=12)
plt.axis('equal');

```
