# Linier Regression with Scikit-Learn 

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data.head()
```

```python
X =data['DIS'].values.reshape(-1,1)
```

```python
Y = data['MDEV'].values
```

```python
Y = Y.reshape(-1,1)
```

```python
# split

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
#import module of score  and accuracy
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
```

```python
# mean squared error

s_train =mean_squared_error(y_train,y_train)
s_test = mean_squared_error(y_test,y_test_pred)

print(f'Mean square error of training set:{s_train}')
print(f'Mean square error of testing set :{s_test}')
```

```python
# r2 score
sr2_train = r2_score(y_train, y_train_pred)
sr2_test = r2_score(y_test,y_test_pred)
print(f'R2 variance score of testing set: {sr2_train}')
print(f'R2 variance score of testing set: {sr2_test}')
```
